```python
file = open('./data/atis/test/label', 'r')
file2 = open('./data/atis/test/label_one', 'w+')
file3 = open('./data/atis/test/label_two', 'w+')

for i,line in enumerate(file):
    line = line.rstrip('\r\n')
    print(line)
    for index, word in enumerate(line):
        if word == '#':
            line = line[0:index]
    words = line.split('_')
    if len(words)==2:
        if words[0] == '':
            words[0] = 'none'
        if words[1] == '':
            words[1] = 'none'
        print(words)
        file2.write(words[0]+'\n')
        file3.write(words[1]+'\n')
    if len(words)==3:
        if words[0] == '':
            words[0] = 'none'
        if words[1] == '':
            words[1] = 'none'
        if words[2] == '':
            words[2] = 'none'
        print(words)
        file2.write(words[0]+'\n')
        file3.write(words[1]+'_'+words[2]+'\n')
```
